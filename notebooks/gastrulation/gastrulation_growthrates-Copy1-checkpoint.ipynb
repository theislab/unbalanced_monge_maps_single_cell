```python
import matplotlib.pyplot as plt
import anndata as ad
import numpy as np
import pandas as pd
import scanpy as sc
import scipy as sp
import seaborn as sb
from moscot.problems.time import TemporalProblem
```

```python
new_adata = sc.read("/lustre/groups/ml01/workspace/monge_velo/data/benchmarks/gastrulation/adata_gastrulation.h5ad")
sc.pp.log1p(new_adata)
new_adata
```

```python
map_dict = {"E6.5": 6.5, "E6.75": 6.75, "E7.0": 7.0, "E7.25": 7.25, "E7.5": 7.5, "E7.75": 7.75, "E8.0": 8.0,
            "E8.25": 8.25, "E8.5": 8.5}
new_adata.obs["stage"] = new_adata.obs["stage"].copy().map(map_dict)
```

```python
leave_out = [6.5, 6.75, 7.0, 7.25, 7.5, 7.75, 8.0]
new_adata = new_adata[~new_adata.obs["stage"].isin(leave_out)].copy()
```

```python
tp = TemporalProblem(new_adata)
tp = tp.prepare("stage")
tp = tp.score_genes_for_marginals(gene_set_proliferation="mouse", gene_set_apoptosis="mouse", proliferation_key="prolif")
```

```python
sc.pl.umap(new_adata[new_adata.obs["stage"] == 8.5], color=['prolif', "apoptosis"])
```

```python
new_adata.obs["growth_rate"] = np.exp(new_adata.obs["prolif"] - new_adata.obs["apoptosis"])
new_adata.obs["scaled_growth_rate"] = np.exp(2 * (new_adata.obs["prolif"] - new_adata.obs["apoptosis"]))
new_adata.obs["scaled3_growth_rate"] = np.exp(3 * (new_adata.obs["prolif"] - new_adata.obs["apoptosis"]))
new_adata.obs["scaled4_growth_rate"] = np.exp(4 * (new_adata.obs["prolif"] - new_adata.obs["apoptosis"]))
```

```python
sc.pl.umap(new_adata[new_adata.obs["stage"] == 8.5], color=["growth_rate", "scaled_growth_rate"])
```

```python
sc.pl.umap(new_adata[new_adata.obs["stage"] == 8.5], color=["scaled3_growth_rate", "scaled4_growth_rate"])
```
