```python
from moscot.problems.time import TemporalNeuralProblem
```

```python
org_adata = sc.read("/lustre/groups/ml01/workspace/monge_velo/data/benchmarks/larry_invitro/larry_invitro_adata_sub_raw.h5ad")

leave_out = [2.,]
adata = org_adata[~org_adata.obs['time_info'].isin(leave_out)].copy()

adata.obs = adata.obs.loc[:, ['time_info', 'state_info']]
adata.var = adata.var[[]]
adata.uns = {}

sc.pp.log1p(adata)

sc.pp.pca(adata)
```

```python
adata
```

```python
tnp = TemporalNeuralProblem(adata)
tnp = tnp.score_genes_for_marginals("mouse", "mouse")
```

```python
tnp = tnp.prepare("time_info", joint_attr="X_pca")
tnp = tnp.solve(tau_a=0.95, tau_b=0.95)
```

```python
with open("/lustre/groups/ml01/workspace/monge_velo/data/benchmarks/larry_invitro/test_save_tnp", "wb") as f:
    cloudpickle.dump(tnp, f)

tnp
```
